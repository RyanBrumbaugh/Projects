```python
def plotscatter3Ddata(fit_x, fit_y, fit_z, scat_x, scat_y, scat_z):
  fig = plt.figure(figsize=(30,20))
  ax1 = fig.add_subplot(2, 2, 1, projection='3d') #Two rows of two columns
  ax1.scatter3D(scat_x, scat_y, scat_z)
  ax1.plot3D(fit_x.flatten(), fit_y.flatten(), fit_z.flatten(), 'blue')
  ax1.view_init(0, 90) ########################
  ax1.set_xlabel('x', color='red', fontsize = 20)
  ax1.set_ylabel('y', color='red', fontsize = 20)
  ax1.set_zlabel('z', color='red', fontsize = 20)

  ax2 = fig.add_subplot(2, 2, 2, projection='3d') #Two rows of two columns
  ax2.scatter3D(scat_x, scat_y, scat_z)
  ax2.plot3D(fit_x.flatten(), fit_y.flatten(), fit_z.flatten(), 'blue')
  ax2.view_init(40, 0) ########################
  ax2.set_xlabel('x', color='red', fontsize = 20)
  ax2.set_ylabel('y', color='red', fontsize = 20)
  ax2.set_zlabel('z', color='red', fontsize = 20)

  ax3 = fig.add_subplot(2, 2, 3, projection='3d') #Two rows of two columns
  ax3.scatter3D(scat_x, scat_y, scat_z)
  ax3.plot3D(fit_x.flatten(), fit_y.flatten(), fit_z.flatten(), 'blue')
  ax3.view_init(40, 90) ########################
  ax3.set_xlabel('x', color='red', fontsize = 20)
  ax3.set_ylabel('y', color='red', fontsize = 20)
  ax3.set_zlabel('z', color='red', fontsize = 20)

  ax4 = fig.add_subplot(2, 2, 4, projection='3d') #Two rows of two columns
  ax4.scatter3D(scat_x, scat_y, scat_z)
  ax4.plot3D(fit_x.flatten(), fit_y.flatten(), fit_z.flatten(), 'blue')
  ax4.view_init(-90, 0) ########################
  ax4.set_xlabel('x', color='red', fontsize = 20)
  ax4.set_ylabel('y', color='red', fontsize = 20)
  ax4.set_zlabel('z', color='red', fontsize = 20)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential 
from keras.layers import Dense

model = Sequential()
model.add(Dense(100, input_shape=(2,))) #expects 2 columns
model.add(Dense(200, activation='relu'))
model.add(Dense(300, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(300, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.compile(loss='mean_squared_error', 
              optimizer='SGD')
```

```python
model.fit(trained_X, trained_y,epochs=200)
```

```python
y_pred = model.predict(trained_X)
```
